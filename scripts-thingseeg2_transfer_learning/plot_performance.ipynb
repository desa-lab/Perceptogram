```python
transfered= 'transferred'
transfer_source='other'
sub = 1
performances_transferred = []
for common_size in range(500,5500,500):
    param=f'{common_size}avg_{transfered}_from{transfer_source}'
    performances_dir = f'../results/thingseeg2_preproc/sub-{sub:02d}/performances_versatile_diffusion_{param}.npy'
    performances = np.load(performances_dir)
    performances_transferred.append(performances[5])
# transfered= 'nontransferred'
# performances_nontransferred = []
# for common_size in range(500,5500,500):
#     param=f'{common_size}avg10_{transfered}_from{transfer_source}'
#     performances_dir = f'../results/thingseeg2_preproc/sub-{sub:02d}/performances_versatile_diffusion_{param}.npy'
#     performances = np.load(performances_dir)
#     performances_nontransferred.append(performances[5])
performances_s1_alone = []
for common_size in range(500,5500,500):
    param=f'{common_size}avg'
    performances_dir = f'../results/thingseeg2_preproc/sub-{sub:02d}/performances_versatile_diffusion_{param}.npy'
    performances = np.load(performances_dir)
    performances_s1_alone.append(performances[5])
plt.figure(figsize=(4,4))
plt.plot(range(500,5500,500),performances_transferred,label='augmented')
# plt.plot(range(500,5500,500),performances_nontransferred,label='naive')
plt.plot(range(500,5500,500),performances_s1_alone,label='s1 alone')
plt.xlabel('train size')
plt.ylabel('CLIP performance')
plt.legend()
```
